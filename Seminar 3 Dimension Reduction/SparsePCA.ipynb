 # Comparison of Sparse PCA with PCA
 
 Comparison of sparse PCA with regular PCA on Friedman #1 data
 
 By Yaohang Li

```python
import numpy as np
from sklearn.datasets import make_friedman1
from sklearn.decomposition import SparsePCA
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

## Preparation of Friedman # 1 data

```python
x = np.arange(0, 30)
X, _ = make_friedman1(n_samples=200, n_features=30, random_state=0)
```

## PCA transform

```python
transformer = PCA(n_components=5, random_state=0)
transformer.fit(X)
```

```python
X_transformed = transformer.transform(X)
X_transformed.shape
```

## Contribution of each features to the principle component vecctors

```python
plt.plot(x, transformer.components_[0], label='component 0')
plt.plot(x, transformer.components_[1], label='component 1')
plt.plot(x, transformer.components_[2], label='component 2')
plt.plot(x, transformer.components_[3], label='component 3')
plt.plot(x, transformer.components_[4], label='component 4')
plt.legend(loc='best', bbox_to_anchor=(1, 0.5))
plt.title('PCA')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_transformed[:,0], X_transformed[:,1], X_transformed[:,2], c='r')
plt.show()
```

# Now try Sparse PCA

```python
X, _ = make_friedman1(n_samples=200, n_features=30, random_state=0)
transformer = SparsePCA(n_components=5, alpha=0.5, random_state=0)
transformer.fit(X)
```

```python
X_transformed = transformer.transform(X)
X_transformed.shape
```

## Contribution of each features to the principle component vecctors

```python
plt.plot(x, transformer.components_[0], label='component 0')
plt.plot(x, transformer.components_[1], label='component 1')
plt.plot(x, transformer.components_[2], label='component 2')
plt.plot(x, transformer.components_[3], label='component 3')
plt.plot(x, transformer.components_[4], label='component 4')
plt.legend(loc='best', bbox_to_anchor=(1, 0.5))
plt.title('Sparse PCA')
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X_transformed[:,0], X_transformed[:,1], X_transformed[:,2], c='r')
plt.show()
```
