# Dimension Reduction Demo
Show the true dimensionality of an image

By Yaohang Li

```python
import matplotlib.pyplot as plt
from matplotlib import image
from PIL import Image 
import numpy as np
```

```python
maxiter = 100
```

## load the image

```python
A = image.imread('baboon.gif')
```

```python
n1, n2 = A.shape
```

```python
plt.imshow(A, cmap='gray')
plt.xticks(())
plt.yticks(())
plt.show()
```

## Obtain the low dimensional represention
k is the dimensionaltiy

```python
k = 1
```

```python
x = np.random.rand(n1, k)
y = np.random.rand(n2, k)
```

```python
for _ in range(maxiter):
    x = A.dot(y).dot(np.linalg.inv(y.T.dot(y)))
    y = A.T.dot(x).dot(np.linalg.inv(x.T.dot(x)))
```

```python
plt.imshow(x.dot(y.T), cmap='gray')
plt.xticks(())
plt.yticks(())
plt.show()
```
